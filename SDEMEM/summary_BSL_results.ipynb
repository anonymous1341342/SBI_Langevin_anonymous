```python
from utils_SDEMEM import *
import os
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
import sys
import json
import pandas as pd
from pathlib import Path
import torch
import ot
import numpy as np
import matplotlib.pyplot as plt
```

```python
# ===== Setting ===== #
obs_size = 200
T = 30
theta_dim = 12
x_dim = 180

prior_mean = torch.tensor([5, 1, 3, -1.5, -0.694, -3, 0.027, 0, -0.8, -0.8, -0.8, -0.8], dtype=torch.float32)
prior_std = torch.tensor([1, 1, 1, 1, 0.6, 0.5, 1, 1, 0.5, 0.5, 0.5, 0.5], dtype=torch.float32)

# [log_m0, log_scale, log_offset, log_sigma, mu_delta, mu_gamma, mu_k, mu_t0, log_tau_delta, log_tau_gamma, log_tau_k, log_tau_t0]
theta_true = torch.tensor([5.7, 0.7, 2.08, -1.6, -0.694, -3, 0.027, 0, -1.15, -1.15, -1.15, -1.15], dtype=torch.float32).reshape(1, -1)
```

```python
task_id = 0
chain_id = 0

theta_BSL = torch.from_numpy(np.load(f'res_BSL_plain_corr/theta_BSL_Task{task_id}_chain{chain_id}.npy'))
```

```python
task_id = 0
chain_id = 0

theta_BSL_starttruth = torch.from_numpy(np.load(f'res_BSL_plain_corr_starttruth/theta_BSL_Task{task_id}_chain{chain_id}.npy'))
```

```python
var_name = [
    r"$\log m_0$",
    r"$\log \mathrm{scale}$",
    r"$\log \mathrm{offset}$",
    r"$\log \sigma$",
    r"$\mu_{\delta}$",
    r"$\mu_{\gamma}$",
    r"$\mu_{k}$",
    r"$\mu_{t_0}$",
    r"$\log \tau_{\delta}$",
    r"$\log \tau_{\gamma}$",
    r"$\log \tau_{k}$",
    r"$\log \tau_{t_0}$"
]
```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

data_np = theta_BSL.cpu().numpy()

fig, axes = plt.subplots(2, 5, figsize=(16, 6), sharex=True)
axes = axes.flatten()

# Plot all dimensions except dimensions 0, 1, and 6
plot_dims = [i for i in range(12) if i not in [0, 1, 6]]

for ax, i in zip(axes, plot_dims):
    ax.plot(data_np[:, i])
    ax.set_title(var_name[i], fontsize=14)
    ax.axhline(
        y=theta_true[0, i].item(),
        color="r",
        linestyle="--"
    )
    ax.grid(True)

# Plot the sum of the three
axes[9].plot(data_np[:, 0] + data_np[:, 1] + data_np[:, 6])
axes[9].set_title(r"$\log \mathrm{scale} + \log m_0 + \mu_k$", fontsize=14)
axes[9].axhline(
    y=theta_true[0, 0].item() + theta_true[0, 1].item() + theta_true[0, 6].item(),
    color="r",
    linestyle="--"
)
axes[9].grid(True)


fig.suptitle(
    "BSL trace plot initializing at the localization mean",
    fontsize=20,
    y=0.96
)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig("figs/BSL_trace_plot.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
